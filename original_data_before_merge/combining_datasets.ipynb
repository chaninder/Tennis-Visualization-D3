```python
import pandas as pd
import numpy as np
```

## Read in datasets

```python
matches = pd.read_csv('Match.csv')
matches.head()
```

```python
players = pd.read_csv('Player.csv')
players.head()
```

```python
stats = pd.read_csv('Stats.csv')
stats.head()
```

## Combine datasets

```python
m1 = players.merge(stats, on='player_id', how='inner')
m2 = m1.merge(matches, on='match_id', how='inner')
```

```python
final = m2.drop(columns=[
    'player_id', 'max_odds', 'bp_faced', '1', '2', '3', '4', '5', 'avg_odds', 'match_id', 'pts'
], axis=1).reset_index().drop('index', axis=1)

final = final.dropna()
```

```python
#final.to_csv('data.csv')
```

```python
YEAR_OF_CHOICE = 2010
final = m2[m2['year'] == YEAR_OF_CHOICE]

final = final.drop(columns=[
    'player_id', 'max_odds', 'bp_faced', '1', '2', '3', '4', '5', 'avg_odds', 'first_serve_per'
], axis=1).reset_index().drop('index', axis=1)
```
